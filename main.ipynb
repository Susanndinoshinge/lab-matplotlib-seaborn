#### Use plt.subplots(nrows=1, ncols=2) to create the plot.

```python
# Method 2
#Using plt.subplots()
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

axes[0].plot(x, y, label='y = 2x', color='blue', linestyle='-', linewidth=2)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('Line Plot: y = 2x')
axes[0].legend()
axes[0].grid(True)

axes[1].plot(x, z, label='z = x^2', color='red', linestyle='--', linewidth=2)
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
axes[1].set_title('Line Plot: z = x^2')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

#### Plot both $y=x^2$ and $y=e^x$ in the same plot using normal and logarithmic scale.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
x = np.linspace(0, 3, 100) 
y1 = x**2
y2 = np.exp(x)
```

```python
# Plot using Normal scale 

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(x, y1, label=r'$y=x^2$', color='blue')
plt.plot(x, y2, label=r'$y=e^x$', color='green')
plt.title('Normal Scale')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot using logarithmic scale
plt.subplot(1, 2, 2)
plt.plot(x, y1, label=r'$y=x^2$', color='blue')
plt.plot(x, y2, label=r'$y=e^x$', color='green')
plt.title('Logarithmic Scale')
plt.yscale('log') 
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python


fig, axes = plt.subplots(1, 2, figsize=(12, 4))
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
# Plot 1: Median steps by month

fitbit.groupby('Months_encoded')['Steps'].median().plot(kind='bar', color='purple', ax=axes[0])
axes[0].set_title('Median Steps by Month_Encoded')
axes[0].set_xlabel('Month')
axes[0].set_ylabel('Median Steps')
axes[0].tick_params(axis='x', rotation=0)



# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
# Plot 2: Median steps by weekday/weekend

median_steps_by_daytype = fitbit.groupby('Work_or_Weekend')['Steps'].median()
median_steps_by_daytype.index = ['Work' if idx == 0 else 'Weekend' for idx in median_steps_by_daytype.index]

colors = ['orange' if day == 'Work' else 'pink' for day in median_steps_by_daytype.index]

median_steps_by_daytype.plot(kind='bar', color=colors, ax=axes[1])
axes[1].set_title('Median Steps by Work/Weekend')
axes[1].set_xlabel('Day Type')
axes[1].set_ylabel('Median Steps')
axes[1].set_xticks(range(len(median_steps_by_daytype.index)))
axes[1].set_xticklabels(median_steps_by_daytype.index, rotation=0)


plt.tight_layout()
plt.savefig('figures/fitbit_visualization.png')
plt.show()

```

#### Use the right visulalization to show the distribution of column `Age`.

```python
# your code here
#Histogram for the age column
plt.figure(figsize=(8, 5))
plt.hist(titanic['Age'].dropna(), bins=20, color='pink', edgecolor='purple')
plt.title('Distribution of Age in Titanic Dataset')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

#### Use subplots and plot the distribution of the `Age`  with bins equal to 10, 20 and 50.

```python
# your code here
bin_sizes = [10, 20, 50]
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)

for i, bins in enumerate(bin_sizes):
    axes[i].hist(titanic['Age'].dropna(), bins=bins, color='purple', edgecolor='pink')
    axes[i].set_title(f'Distribution of Age (Bins={bins})')
    axes[i].set_xlabel('Age')
    axes[i].set_ylabel('Frequency')
    axes[i].grid(True)

plt.tight_layout()
```

#### How does the bin size affect your plot?

```python
"""
Bin Size = 10:
Effect: Fine-grained details are revealed in the distribution.
Observation: Peaks and variations in the age distribution are more pronounced, allowing for a detailed view.

Bin Size = 20:
Effect: Strikes a balance between detail and generalization.
Observation: The histogram provides a moderate level of detail, highlighting patterns without excessive noise.

Bin Size = 50:
Effect: Broadens the overview, smoothing out variations.

Observation: The histogram gives a broader, smoothed representation of the age distribution, potentially overlooking finer details.
In summary, the choice of bin size influences the level of detail in the visualization. Smaller bins reveal more specifics, larger bins provide a broader overview,
and an optimal bin size balances both, capturing meaningful patterns without introducing excessive noise or oversimplifying the distribution. 
Adjust bin size based on your specific dataset characteristics and the insights you seek.
"""
```
